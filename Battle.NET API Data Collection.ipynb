```python

# Set the API endpoint for the token endpoint
token_endpoint = "https://us.battle.net/oauth/token"

# Set the required client credentials
client_id = "HIDDEN"
client_secret = "HIDDEN"

# Set the request header to include the required content type
header = {
    "Content-Type": "application/x-www-form-urlencoded"
}

# Set the request payload to include the client credentials and authorization grant type
payload = {
    "grant_type": "client_credentials",
    "client_id": client_id,
    "client_secret": client_secret
}

# Make the POST request to the token endpoint
response = requests.post(token_endpoint, headers=header, data=payload)

# Get the access token from the response
access_token = response.json().get("access_token")

access_token
```

```python
# PLAYER DATABASE TABLE

all_players = response_json['players']

players_dict = {
    "player_id": [],
    "given_name": [],
    "familyName": [],
    "roles":[],
    "TeamID":[],
    "competitions" : []
    }    

for player_id in all_players:
    players_dict["player_id"].append(player_id)
    players_dict["given_name"].append(all_players[player_id]['givenName'])
    players_dict["familyName"].append(all_players[player_id]['familyName'])
    players_dict['roles'].append(all_players[player_id]['role'])
    players_dict['TeamID'].append(all_players[player_id]['currentTeam'])
    #players_dict['competitions'].append(all_players[player_id]['competitions'][0])

    comp2 = (all_players[player_id]['competitions'])
    competition_text = ', '.join(comp2)
    players_dict['competitions'].append(competition_text)


players_dict

```

```python
df = pd.DataFrame(players_dict)
```

```python
engine = create_engine("mysql+pymysql://HIDDEN")
```

```python
df.to_sql(name='players', con=engine, if_exists='append', index=False)
```

```python
# Teams DATABASE TABLE

all_teams = response_json['teams']

teams_dict = {
    "TeamID": [],
    "name": [],
    "code": [],
    "logo" : [],
    "icon" : [],
    "competitions" : [],
    "roster" : [],
    "primaryColor" : [],
    "secondaryColor" : []

    }    

for team_id in all_teams:
    teams_dict["TeamID"].append(team_id)
    teams_dict["name"].append(all_teams[team_id]['name'])
    teams_dict["code"].append(all_teams[team_id]['code'])
    teams_dict["logo"].append(all_teams[team_id]['logo'])
    teams_dict["icon"].append(all_teams[team_id]['icon'])
    teams_dict["primaryColor"].append(all_teams[team_id]['primaryColor'])
    teams_dict["secondaryColor"].append(all_teams[team_id]['secondaryColor'])



    comp2 = (all_teams[team_id]['competitions'])
    competition_text = ', '.join(comp2)
    teams_dict['competitions'].append(competition_text)

    rost = (all_teams[team_id]['roster'])
    roster_text = ''
    for player_id in rost:
        roster_text += str(player_id) + ', '
    teams_dict['roster'].append(roster_text.strip())

teams_dict
```

```python
df_teams = pd.DataFrame(teams_dict)
```

```python
df_teams.to_sql(name='Teams', con=engine, if_exists='append', index=False)
```
